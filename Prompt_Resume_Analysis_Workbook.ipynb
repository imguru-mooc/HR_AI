# 📖 프롬프트 최적화 Jupyter 실습 워크북 – 이력서 분석


## ✅ Step 1. 단순 프롬프트 (실패 예시)

```plaintext
이력서를 분석해줘.
```

❌ 출력 → "이력서는 직업을 찾기 위해 작성하는 문서입니다."



## ✅ Step 2. 대상만 명시 (불완전 개선)

```plaintext
다음 이력서를 분석해줘.

이력서:
[이력서 내용 입력]
```

✔️ 출력 → 단순 요약만 가능.  
강점/약점 도출 불가.  



## ✅ Step 3. 목적 명시 (상당히 개선)

```plaintext
다음 이력서를 기반으로 요약과 강점 3가지, 약점 2가지를 알려줘.

이력서:
[이력서 내용 입력]
```

✔️ 출력 → 요약 + 강점/약점 도출 가능.



## ✅ Step 4. 역할 부여 + 포맷 명시 (전문가 수준)

```plaintext
너는 전문 헤드헌터입니다.  
다음 이력서를 분석해 면접관에게 제공할 보고서를 작성해줘.  
결과는 Markdown 포맷으로.

포함 내용:
- 요약
- 강점 3가지
- 약점 2가지
- 추천 면접 질문 3개

이력서:
[이력서 내용 입력]
```

✔️ 출력 → 구조화된 전문가용 보고서.



## ✅ Step 5. 확장형 프롬프트 (AI 심화 분석)

```plaintext
다음 이력서를 분석해서 직무 적합도(상, 중, 하) 평가와 성장 가능성을 함께 포함한 전문가 보고서를 Markdown 포맷으로 작성해줘.
```

✔️ 출력 → 직무 적합도 + 성장 가능성 + 종합 평가까지 포함.



## 🧾 샘플 이력서 – 데이터 분석 지원자

```markdown
## 📍 개인 정보
- 이름: 이지원
- 연락처: 010-XXXX-XXXX
- 이메일: leejiwon@example.com
- 주소: 서울특별시 서초구 ○○로 ○○

## 💼 경력 요약
- 데이터 분석가 | XYZ 커머스 (2021.05 ~ 현재)
  - 이커머스 고객 이탈 예측 모델 개발
  - 광고 성과 데이터 분석 및 ROAS 최적화
  - Tableau 대시보드 제작 및 지표 모니터링
  - SQL, Python을 활용한 데이터 ETL 및 자동화

- 마케팅 인턴 | ABC 디지털 마케팅 (2020.03 ~ 2020.08)
  - SNS 캠페인 데이터 분석
  - 고객 세분화 및 타겟팅 전략 지원

## 🚀 주요 프로젝트
1. 고객 이탈 예측 모델 구축
   - 랜덤 포레스트 기반 예측 모델
   - 이탈률 15% → 10% 감소
2. 광고 성과 최적화 분석
   - Facebook/Google 광고 데이터 분석
   - ROAS 25% 향상 기여
3. 고객 행동 데이터 대시보드 개발
   - Tableau로 구매 전환 경로 분석 대시보드 설계

## 🔧 보유 기술
| 분야        | 기술                                          |
|-------------|-----------------------------------------------|
| 프로그래밍  | Python (pandas, scikit-learn), SQL           |
| 시각화 도구 | Tableau, PowerBI                             |
| 데이터 엔진 | MySQL, BigQuery                              |
| 분석 기법   | 시계열 분석, 머신러닝 분류 모델, 군집화 분석 |
| 협업 도구   | Jira, Notion, Slack                           |

## 🎓 학력
- ○○대학교 경영학과 졸업 (2016.03 ~ 2021.02)

## 🏅 자격증
- SQLD (SQL 개발자)
- 데이터 분석 준전문가 (ADsP)

## 🌍 언어
- 한국어: 원어민
- 영어: 비즈니스 회화 가능 (TOEIC 870)
```
