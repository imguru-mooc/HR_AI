# 🎯 생성형 AI로 인사이트 도출 방법

## 🏗️ 전체 프로세스
```
1. 문제 정의
2. 프롬프트 설계
3. 응답 분석 및 재질문
4. 아이디어 구체화 및 정제
5. 검증 및 실행 가능성 평가
6. 실행 계획 수립 및 적용
```


## ✅ STEP 1. 문제 정의
- 목적: 무엇을 해결하고 싶은가?
- 문제: 현재 겪고 있는 구체적인 상황과 문제는?
- 핵심 질문: 인사이트가 필요한 핵심 질문은 무엇인가?

🔍 **예시:**
> "우리 회사 교육 프로그램의 참여율이 낮다. 이를 개선할 방법은?"


## ✅ STEP 2. 프롬프트 설계
- 구체적이고 명확한 프롬프트 작성
- 맥락과 배경 정보를 포함

### ✔️ 프롬프트 유형
- 문제 해결형:
> "직원 교육 참여율을 높이기 위한 창의적인 방법 10가지를 제안해 주세요."
- 비교 분석형:
> "온보딩 교육과 멘토링 프로그램 중 어떤 것이 신입 직원에게 더 효과적일까요?"
- 확장 탐색형:
> "기업 교육 트렌드 중 활용도가 높은 AI 기반 도구는 무엇인가요?"


## ✅ STEP 3. 응답 분석 및 재질문
- 초반 응답을 바탕으로 더 깊이 파고들 질문을 설계
- 다각도 질문 시도

🔁 **반복 프롬프트 예시:**
> "이 중에서 실행 비용이 낮고 효과가 높은 3가지를 추천해 주세요."
> "이 방법을 도입할 때 예상되는 리스크는 무엇인가요?"


## ✅ STEP 4. 아이디어 구체화 및 정제
- 도출된 아이디어를 실행 가능한 형태로 전환
- 필요시 SWOT 분석, 5 WHY 분석 활용

🛠️ **실행 형태 변환:**
> "게이미피케이션 교육 → 어떻게 적용할 것인지 단계별 실행안으로 바꿔줘."


## ✅ STEP 5. 검증 및 실행 가능성 평가

### ✔️ IDEA Validation Canvas
| 항목 | 질문 | 검토 포인트 |
|------|------|--------------|
| **Impact** | 얼마나 큰 효과가 있는가? | 생산성 향상, 몰입도 증가 |
| **Desirability** | 직원들이 원하는가? | 참여도, 만족도 |
| **Feasibility** | 실행 가능한가? | 예산, 기술, 시간 |
| **Viability** | 지속 가능한가? | 유지보수, 장기적 ROI |


## ✅ STEP 6. 실행 계획 수립 및 적용
- 우선순위 결정
- 로드맵 작성
- 프로토타입 테스트 → 전사 확산

🚀 **실행 예시:**
| 단계 | 활동 | 기간 | 담당 부서 |
|------|------|------|------------|
| 설계 | 교육 콘텐츠 설계 | 1주 | 교육팀 |
| 개발 | 플랫폼 적용 및 도구 개발 | 2주 | IT팀 |
| 테스트 | 시범 운영 | 1주 | 교육팀 |
| 확산 | 전체 적용 | 1주 | 교육팀 |


## 📌 Tip: 생성형 AI 활용 공식
```
[문제 정의] + [배경 정보] + [조건] → ChatGPT 프롬프트 설계 → 응답 분석 → 재질문 → 실행 전략화
```
