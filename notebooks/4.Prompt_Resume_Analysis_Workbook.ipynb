# 📖 프롬프트 최적화 실습 워크북 – 이력서 분석


## ✅ Step 1. 단순 프롬프트 (실패 예시)

```plaintext
이력서를 분석해줘.
```

❌ 출력 → "이력서는 직업을 찾기 위해 작성하는 문서입니다."



## ✅ Step 2. 대상만 명시 (불완전 개선)

```plaintext
다음 이력서를 분석해줘.

이력서:
[이력서 내용 입력]
```

✔️ 출력 → 단순 요약만 가능.  
강점/약점 도출 불가.  



## ✅ Step 3. 목적 명시 (상당히 개선)

```plaintext
다음 이력서를 기반으로 요약과 강점 2가지, 약점 2가지를 알려줘.

이력서:
[이력서 내용 입력]
```

✔️ 출력 → 요약 + 강점/약점 도출 가능.



## ✅ Step 4. 역할 부여 + 포맷 명시 (전문가 수준)

```plaintext
너는 전문 헤드헌터입니다.  
다음 이력서를 분석해 면접관에게 제공할 보고서를 작성해줘.  
결과는 Markdown 포맷으로.

포함 내용:
- 요약
- 강점 3가지
- 약점 2가지
- 추천 면접 질문 3개

이력서:
[이력서 내용 입력]
```

✔️ 출력 → 구조화된 전문가용 보고서.



## ✅ Step 5. 확장형 프롬프트 (AI 심화 분석)

```plaintext
다음 이력서를 분석해서 직무 적합도(상, 중, 하) 평가와 성장 가능성을 함께 포함한 전문가 보고서를 Markdown 포맷으로 작성해줘.
```

✔️ 출력 → 직무 적합도 + 성장 가능성 + 종합 평가까지 포함.
