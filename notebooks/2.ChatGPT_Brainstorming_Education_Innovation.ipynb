# 🧠 ChatGPT 브레인스토밍 실습
## 🎯 주제: 교육 프로그램 혁신 아이디어 도출 + 단계적 프롬프트 + 검증 프레임워크


## ✅ 실습 목표
- ChatGPT를 활용해 교육 프로그램 혁신 아이디어를 도출한다.
- 프롬프트를 단계적으로 설계 및 개선하며 효과적으로 질문하는 법을 학습한다.
- 도출된 아이디어를 체계적으로 평가하고 실행 가능성을 검증한다.


## 🗺️ 실습 전체 흐름
```
1. 문제 정의
2. 초기 프롬프트 작성
3. 응답 검토 및 개선
4. 세부 질문 확장
5. 최종 아이디어 도출
6. 검증 및 우선순위 설정
```


# 🔥 STEP 1: 문제 정의
- 교육 참여율이 낮다.
- 교육이 실무와 연결되지 않는다.
- 세대별(MZ세대) 교육 몰입도가 부족하다.
- 교육 효과 측정 및 피드백이 미흡하다.


# 🚀 STEP 2: 프롬프트 단계적 설계
## ✔️ 1단계: 기본 탐색형 프롬프트
> 우리 회사의 교육 프로그램을 더 효과적이고 혁신적으로 만들기 위한 아이디어 10가지를 제안해 주세요.
## ✔️ 2단계: 회사 상황을 명시한 프롬프트
> 우리 회사는 디지털 트랜스포메이션 시대에 맞춰 MZ세대 직원의 학습 몰입도와 실무 연결성을 높이고자 합니다. 
> 효과적이고 몰입감 높은 교육 프로그램 혁신 아이디어 10가지를 구체적으로 제안해 주세요.
## ✔️ 3단계: 특정 문제 집중 프롬프트 (세부화)
- 참여율 향상:
> 직원들의 교육 참여율을 높일 수 있는 창의적인 방법은 무엇인가요?
- 학습 몰입도:
> MZ세대 직원들이 몰입할 수 있는 교육 콘텐츠나 학습 방식은 무엇인가요?
- 실무 연계:
> 교육 내용이 업무와 직결되도록 하는 방법은 무엇인가요?
- 교육 효과 측정:
> 교육 효과를 데이터로 측정하고 피드백할 수 있는 방법은 무엇인가요?


## ✔️ 4단계: 리스크 분석 및 보완 프롬프트
> 이 교육 프로그램 혁신 아이디어를 실행할 때 예상되는 리스크는 무엇이며, 이를 어떻게 해결할 수 있나요?


## ✔️ 5단계: 실행 전략 구체화 프롬프트
> 이 교육 혁신 아이디어를 실제로 도입하기 위한 단계별 실행 계획을 작성해 주세요. 
> 각 단계별 주요 활동과 예상 소요 시간도 포함해 주세요.


# 🔍 STEP 4: 검증 프레임워크 적용
## ✔️ IDEA Validation Canvas
| 항목 | 질문 | 검증 포인트 |
|------|------|-------------|
| Impact | 교육 효과와 업무 성과에 얼마나 기여하는가? | 성과 향상, 생산성, 역량 강화 |
| Desirability | 직원들이 원하고 몰입할 수 있는가? | 참여율, 만족도 |
| Feasibility | 현재 자원과 인프라로 실행 가능한가? | 예산, 시간, 기술 |
| Viability | 장기적으로 유지 및 발전 가능한가? | 유지보수, 확장성, ROI |


## ✔️ 우선순위 매트릭스
|                | 실현 가능성 낮음 | 실현 가능성 높음 |
|----------------|-------------------|-------------------|
| 효과 낮음      | ❌ 제외            | 🔍 재검토          |
| 효과 높음      | 🤔 도전 과제       | ✅ 최우선 실행     |


# 🏁 STEP 5: 최종 실행 계획 작성
| 단계 | 활동 | 예상 기간 | 담당 부서 |
|------|------|-----------|------------|
| 아이디어 선정 | IDEA Canvas 평가 | 1일 | 교육팀 |
| 벤치마킹 | 우수 사례 조사 | 1주 | 교육팀 + IT팀 |
| 설계 및 개발 | 교육 플랫폼 설계 및 개발 | 2~3주 | IT팀 |
| 시범 테스트 | 일부 팀 대상 테스트 | 1주 | 교육팀 |
| 전사 도입 | 전체 적용 및 개선 반영 | 1주 | 교육팀 + IT |
