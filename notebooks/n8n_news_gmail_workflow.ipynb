
# 🚀 n8n 자동화 실습 노트북 (v1.99.0 기준)

## ✅ n8n 기본 설명

### 🔸 n8n이란?
- No-code / Low-code 자동화 플랫폼
- Zapier, Make와 유사하지만 **오픈소스 + 무제한 워크플로우 실행 가능**
- 클라우드 서비스(n8n.io) 또는 자체 서버 설치 가능

### 🔸 주요 특징
- 700개 이상의 노드 지원 (API, 데이터베이스, 파일, AI, 메신저 등)
- 복잡한 조건 분기 (IF, Switch 등)
- 반복, 함수 처리 (Loop, Code node)
- 데이터 변환, API 연결 자동화

---

## ✅ n8n 가입 방법 (Cloud 기준)

1. 🔗 사이트 접속 → [https://n8n.io](https://n8n.io)
2. **"Sign up"** 클릭
3. 이메일, 비밀번호 입력 후 가입
4. 가입 후 기본 워크스페이스 생성
5. 클라우드 대시보드 진입 → 워크플로우 시작 가능

> 🌟 참고: 무료 요금제도 존재하며, 제한은 있지만 개인 실습용으로 충분

## ✅ 워크플로우 설계도

```
Start → HTTP Request (뉴스 검색) → Gmail (이메일 전송)
```

---

## ✅ 단계별 실습

### ✔️ Step 1: 워크플로우 생성
- 대시보드 → **New Workflow** 클릭

### ✔️ Step 2: HTTP Request 노드 (뉴스 검색)
- **노드 추가 → HTTP Request 선택**
- 설정:
    - **Method:** GET
    - **URL:** https://news.google.com/rss/search?q=인공지능
    - **Response Format:** XML
- **Execute Node** 클릭 → 데이터 미리보기

### ✔️ Step 3: RSS → JSON 변환
- 응답은 자동으로 JSON 변환됨
- 구조 예시:

```json
{
  "rss": {
    "channel": [
      {
        "item": [
          { "title": "기사 제목", "link": "링크", "description": "설명" }
        ]
      }
    ]
  }
}
```

### ✔️ Step 4: Gmail 노드 연결
- **노드 추가 → Gmail 선택**
- **Resource:** Message
- **Operation:** Send
- 설정:
    - **Recipient:** 본인 이메일
    - **Subject:** `[뉴스] 인공지능 최신 뉴스`
    - **Message:**

```
{{ $json["rss"]["channel"][0]["item"][0]["title"] }}
{{ $json["rss"]["channel"][0]["item"][0]["link"] }}
```

- OAuth2 인증 → Gmail 계정 연결

### ✔️ Step 5: 데이터 포맷팅 (옵션)
- **Code 노드** 활용 예시:

```javascript
const items = $json["rss"]["channel"][0]["item"];
let body = "오늘의 인공지능 뉴스:\n\n";

for (let i = 0; i < Math.min(5, items.length); i++) {
  body += `• ${items[i].title}\n${items[i].link}\n\n`;
}

return [{ json: { body } }];
```

- Gmail 본문 → `{{$json["body"]}}`

## ✅ 체크리스트

| 단계 | 노드 | 역할 |
|------|------|------|
| 1 | Start | 시작 |
| 2 | HTTP Request | 뉴스 데이터 크롤링 (RSS) |
