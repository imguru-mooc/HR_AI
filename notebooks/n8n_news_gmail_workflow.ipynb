# 🚀 n8n 자동화 실습 노트북 (AI Agent 버전, v1.99.0 기준)

## ✅ n8n 기본 설명

### 🔸 n8n이란?
- No-code / Low-code 자동화 플랫폼
- Zapier, Make와 유사하지만 **오픈소스 + 무제한 워크플로우 실행 가능**
- 클라우드 서비스(n8n.io) 또는 자체 서버 설치 가능

### 🔸 주요 특징
- 700개 이상의 노드 지원 (API, 데이터베이스, 파일, AI, 메신저 등)
- 복잡한 조건 분기 (IF, Switch 등)
- 반복, 함수 처리 (Loop, AI Agent node)
- 데이터 변환, API 연결 자동화

---

## ✅ n8n 가입 및 시작 방법

1. 🔗 [https://n8n.io](https://n8n.io) 접속
2. 우측 상단 → **Sign up** 클릭
3. 이메일, 이름, 비밀번호 입력 → 계정 생성
4. 이메일 인증 후 대시보드 접속
5. 기본 워크스페이스 생성 → **New Workflow** 클릭 → 워크플로우 시작

> 🌟 무료 플랜: 개인 실습, 테스트 충분

---

## ✅ 실습 목표

> ✅ 인터넷 뉴스 검색 → ✅ AI Agent로 메일 본문 생성 → ✅ Gmail로 전송

---

## ✅ 워크플로우 설계도

```
Start → HTTP Request (뉴스 검색) → AI Agent (본문 생성) → Gmail (메일 전송)
```

---

## ✅ 단계별 실습

### ✔️ Step 1: 워크플로우 생성
- 대시보드 → **New Workflow** 클릭

### ✔️ Step 2: HTTP Request 노드 설정 (뉴스 검색)
- **노드 추가 → HTTP Request**
- 설정:
    - **Method:** GET
    - **URL:** https://news.google.com/rss/search?q=인공지능
    - **Response Format:** XML
- **Execute Node** 클릭 → 뉴스 데이터 미리보기

### ✔️ Step 3: AI Agent 노드 설정 (본문 생성)
- **노드 추가 → AI Agent**
- 설정:
    - **Model:** GPT-4 또는 GPT-3.5
    - **Input:**

```
다음은 뉴스 RSS 데이터입니다. 기사 제목과 링크만 골라서 아래 형식으로 정리해주세요.

형식 예시:
1. 제목 - 링크
2. 제목 - 링크
3. 제목 - 링크

뉴스 데이터:
{{ $json["rss"]["channel"][0]["item"] }}
```

- OpenAI API 연결 필요

### ✔️ Step 4: Gmail 노드 설정 (메일 전송)
- **노드 추가 → Gmail**
- 설정:
    - **Recipient:** 본인 이메일
    - **Subject:** [뉴스] 인공지능 최신 뉴스
    - **Message:**

```
{{$json["result"]}}
```

- OAuth2 인증 → Gmail 계정 연결

## ✅ 전체 워크플로우 연결

```
[Start] → [HTTP Request] → [AI Agent] → [Gmail]
```

---

## ✅ 워크플로우 체크리스트

| 단계 | 노드 | 역할 |
|------|------|------|
| 1 | Start | 시작 |
| 2 | HTTP Request | 뉴스 데이터 크롤링 (RSS) |
| 3 | AI Agent | 뉴스 본문 요약 및 포맷팅 |
| 4 | Gmail | 이메일로 전송 |

## ✅ 확장 아이디어
- 키워드 입력 자동화 (Webhook → 검색어 입력)
- 매일 아침 자동 뉴스 메일 발송 (Cron 노드 활용)
- ChatGPT API와 연동해 뉴스 요약 품질 강화
- Slack, Discord, Telegram으로도 뉴스 보내기 확장 가능
