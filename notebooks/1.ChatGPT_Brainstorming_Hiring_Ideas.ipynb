# 🧠 ChatGPT 브레인스토밍 실습
## 🎯 주제: 채용 개선 아이디어 도출 + 프롬프트 설계 + 검증 프레임워크
---
## ✅ 실습 목적
- ChatGPT를 활용해 채용 개선 인사이트를 창출한다.
- 단계적으로 프롬프트를 수정하며 더 나은 답변을 얻는 법을 익힌다.
- 도출된 아이디어를 평가하고 우선순위를 설정하는 검증 프레임워크를 적용한다.


## 🗺️ 실습 전체 흐름
```
1. 문제 정의
2. 초기 프롬프트 작성
3. 응답 검토 및 개선
4. 세부 질문 확장
5. 최종 아이디어 도출
6. 검증 및 우선순위 설정
```


# 🔥 STEP 1: 문제 정의
- 지원자 경험 부족
- 채용 기간 지연
- 직무 적합도 판단 어려움
- 공정성 문제 발생


# 🚀 STEP 2: 프롬프트 단계적 설계
## ✔️ 1단계: 기본 탐색형 프롬프트
> 우리 회사는 서류 → 면접 → 최종 합격 절차로 채용을 진행합니다. 
> 이 채용 프로세스를 더 빠르고 효율적으로 만들 수 있는 창의적인 아이디어 10가지를 제안해 주세요.


## ✔️ 2단계: 회사 상황을 명시한 프롬프트
> 우리 회사는 스타트업이며 채용 속도가 매우 중요합니다. 
> 채용 과정에서 지원자 경험을 개선하고, 공정성을 유지하면서도 채용 속도를 높일 수 있는 채용 개선 아이디어 10가지를 구체적으로 제안해 주세요.


## ✔️ 3단계: 특정 문제 집중 프롬프트 (세부화)
- 지원자 경험 개선
> 채용 과정에서 지원자의 만족도를 높일 수 있는 디지털 도구나 방법을 제안해 주세요.
- 채용 속도 향상
> AI나 자동화를 활용해 채용 속도를 높일 수 있는 방법은 무엇인가요?
- 공정성 강화
> AI 기반 채용 프로세스에서 공정성을 확보할 수 있는 구체적인 방법은?
- 직무 적합도 향상
> 직무 적합성을 효과적으로 판단할 수 있는 데이터 기반 평가 방법은?


## ✔️ 4단계: 리스크 분석 및 보완 프롬프트
> 이 아이디어를 실행할 때 예상되는 리스크는 무엇인가요? 그리고 이 리스크를 어떻게 해결할 수 있나요?


## ✔️ 5단계: 실행 전략 구체화 프롬프트
> 이 아이디어를 실제로 적용하기 위한 단계별 실행 계획을 작성해 주세요. 각 단계별 주요 활동과 예상 소요 시간도 포함해 주세요.


# 🔍 STEP 4: 검증 프레임워크 적용
## ✔️ IDEA Validation Canvas
| 항목 | 질문 | 검증 포인트 |
|------|------|-------------|
| Impact | 얼마나 큰 문제를 해결하는가? | 생산성, 비용 절감, 만족도 향상 |
| Desirability | 지원자와 내부 직원 모두가 원하는가? | 사용자 경험, 공감도 |
| Feasibility | 현재 자원과 기술로 가능한가? | 기술, 인력, 예산 |
| Viability | 경제적, 운영적으로 지속 가능한가? | 유지보수 용이성, ROI |


## ✔️ 우선순위 매트릭스
|                | 실현 가능성 낮음 | 실현 가능성 높음 |
|----------------|-------------------|-------------------|
| 효과 낮음      | ❌ 제외            | 🔍 재검토          |
| 효과 높음      | 🤔 도전 과제       | ✅ 최우선 실행     |


# 🏁 STEP 5: 최종 실행 계획 작성
| 단계 | 활동 | 예상 기간 | 담당 부서 |
|------|------|-----------|------------|
| 아이디어 선정 | IDEA Canvas 평가 | 1일 | HR팀 |
| 벤치마킹 | 유사 사례 조사 | 1주 | HR팀 + IT팀 |
| 설계 및 개발 | 시스템 설계 및 개발 | 2~3주 | IT팀 |
| 시범 테스트 | 일부 직무 대상 테스트 | 1주 | HR팀 |
| 전사 도입 | 전체 적용 및 피드백 반영 | 1주 | HR + IT |
