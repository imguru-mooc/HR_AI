# 🏆 성과평가 지원 실습
## 🔍 1:1 미팅 분석 프로세스 + ChatGPT 분석 프롬프트 포함 + 샘플 대화 데이터


## ✅ STEP 1. 데이터 수집
### 📑 샘플 1:1 미팅 대화
**김매니저:** “지은 씨, 오늘 시간 내줘서 고마워요. 이번 미팅에서는 최근 업무 진행 상황, 어려움, 그리고 향후 성장 방향에 대해 이야기해보면 좋겠어요.”

**이지은:** “네, 매니저님. 저도 최근에 고민이 좀 있어서 잘 됐어요.”

**김매니저:** “지난달 프로젝트 정말 잘했어요. 고객 피드백도 좋았고요. 다만 회의록 정리가 조금 아쉬웠어요.”

**이지은:** “네, 다음부터 더 명확하게 정리하겠습니다.”

**김매니저:** “요즘 본인 커리어에 대해 어떤 고민이 있어요?”

**이지은:** “데이터 분석 역량을 좀 더 키우고 싶어요.”

**김매니저:** “좋아요. 관련 교육도 알아봐 드릴게요.”

**김매니저:** “그럼 회의록 개선과 데이터 분석 교육 참여 두 가지를 진행해봐요.”

**이지은:** “네, 교육 리스트는 제가 먼저 찾아볼게요.”

**김매니저:** “다음 미팅 때 진행 상황 공유해요. 수고했어요.”

**이지은:** “감사합니다.”


## ✅ STEP 2. ChatGPT AI 분석용 프롬프트 설계

### ✔️ 1단계: 요약 프롬프트
> 다음 대화를 간결하게 요약해 주세요.

### ✔️ 2단계: 감정 분석 프롬프트
> 다음 대화에서 각각의 발화별 감정 상태(긍정, 부정, 중립, 희망, 책임감 등)를 분석해 주세요.
> 또한 전체 대화의 전반적인 감정 상태를 요약해 주세요.

### ✔️ 3단계: 주제 및 키워드 분석 프롬프트
> 다음 대화에서 주요 주제와 핵심 키워드를 추출해 주세요.

### ✔️ 4단계: 대화 밸런스 분석 프롬프트
> 다음 대화에서 리더(김매니저)와 구성원(이지은)의 발화 비율을 분석해 주세요.

### ✔️ 5단계: 문제점 및 우려사항 도출 프롬프트
> 다음 대화를 바탕으로 현재 업무 또는 커뮤니케이션에서 나타나는 문제점과 우려사항을 정리해 주세요.

### ✔️ 6단계: 개선안 제안 프롬프트
> 다음 대화를 바탕으로 리더와 구성원이 더 효과적으로 협업하고 성장할 수 있도록 구체적인 개선안을 제안해 주세요.

### ✔️ 7단계: 실행 로드맵 생성 프롬프트
> 다음 대화에서 도출된 개선안을 바탕으로 실행 계획을 수립해 주세요.


## ✅ STEP 3. 분석 결과 예시
| 항목 | 결과 |
|------|------|
| 감정 분석 | 긍정 70%, 희망 20%, 책임감 10% |
| 대화 밸런스 | 리더 60% / 구성원 40% |
| 주요 주제 | 프로젝트 피드백, 커리어 성장, 업무 조율, 교육 계획 |
| 문제점 | 회의록 품질 부족, 데이터 분석 역량 부족 |
| 개선안 | 회의록 템플릿 개선, 데이터 교육 지원, 업무 우선순위 명확화 |
| 실행 계획 | 회의록 개선(1주), 교육 지원(2주), 업무 조정(1주) |
