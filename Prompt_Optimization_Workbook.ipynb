
# 📖 프롬프트 최적화 워크북 – 이력서 분석 실습용



## ✅ Step 1. 현재 프롬프트 점검

| 항목 | 내 현재 프롬프트 |
|------|-------------------|
| 목적이 명확한가? | ☐ 예 ☐ 아니오 |
| 역할이 명확한가? | ☐ 예 ☐ 아니오 |
| 출력 포맷을 지시했는가? | ☐ 예 ☐ 아니오 |
| 입력 데이터가 충분한가? | ☐ 예 ☐ 아니오 |
| 예상 출력 결과를 명확히 지시했는가? | ☐ 예 ☐ 아니오 |



## 🔍 Step 2. 프롬프트 개선 체크리스트

| 체크리스트                             | 적용 여부 | 메모                              |
|-----------------------------------------|------------|-------------------------------------|
| 🎯 목적을 명확히 작성했다                | ☐          | 예: '면접관에게 제공할 요약 작성' |
| 👤 역할(Role)을 부여했다                | ☐          | 예: '너는 전문 헤드헌터야'        |
| 📑 출력 포맷을 지시했다                 | ☐          | 예: 'Markdown 포맷으로 작성해줘'  |
| 🔍 세부 항목(강점, 약점 등)을 명시했다 | ☐          |                                   |
| 🗂️ 데이터 입력을 충분히 제공했다         | ☐          | Markdown 이력서 포함               |



## ✍️ Step 3. 프롬프트 개선 실습

### 🔸 [초기 프롬프트]  
```plaintext
이력서를 분석해줘.
```

### 🔸 [1차 개선 프롬프트]  
```plaintext
다음 이력서를 분석해서 요약과 강점, 약점을 알려줘.
```

### 🔸 [최종 개선 프롬프트]  
```plaintext
너는 전문 헤드헌터입니다. 아래 이력서를 분석해 면접관에게 제공할 보고서를 작성해줘.  
결과는 Markdown 포맷으로.  

포함 내용:  
- 요약  
- 강점 3가지  
- 약점 2가지  
- 추천 면접 질문 3개  

이력서:  
[이력서 전체 입력]
```



## 🎯 Step 4. 결과 비교

| 프롬프트 버전 | 출력 품질 | 개선 포인트 |
|----------------|-----------|--------------|
| 초기           | ★☆☆☆☆    | 목적 불명확  |
| 1차 개선       | ★★☆☆☆    | 세부 항목 부족 |
| 최종 개선      | ★★★★★    | 명확한 역할, 포맷, 목적 |



## 💡 Step 5. 프롬프트 최종 템플릿

```plaintext
너는 [역할]입니다.  
다음 [데이터]를 기반으로 [목적]을 위한 [출력 포맷] 형태의 결과를 작성해줘.  

포함 내용:  
- [요약]  
- [강점/약점]  
- [추천 질문]  

데이터:  
[이력서, 보고서, 텍스트 등]
```



## 🔥 Step 6. 프롬프트 엔지니어링 공식

```
역할(Role) + 목적(Goal) + 데이터(Data) + 출력 포맷(Output Format) + 세부 지시(Details) = 최적 프롬프트
```
