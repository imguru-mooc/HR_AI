
# 🏆 ChatGPT를 활용한 이력서 분석 실습
## 🚀 프롬프트 엔지니어링 단계별 학습

---

## 📄 1단계: 단순 요청 – 실패하는 프롬프트
**프롬프트**
```
"이력서를 분석해줘."
```
**🔍 문제점**
- 목적 불명확
- 어떤 정보를 분석해야 하는지 모름
- 결과도 애매하거나 엉뚱함

**💡 출력 예시**
```
"이력서는 직업을 찾기 위한 문서입니다."
```

---

## 📄 2단계: 대상 명시 – 약간 개선된 프롬프트
**프롬프트**
```
"다음 이력서를 분석해줘.
- 이름: 김지민
- 경력: 마케팅 5년
- 스킬: SNS 광고, 콘텐츠 기획, 데이터 분석"
```
**🔍 개선 포인트**
- 대상 데이터는 명확해짐
- 하지만 분석 기준 불명확

**💡 출력 예시**
```
"김지민 님은 마케팅 경력이 5년입니다. SNS 광고와 콘텐츠 기획, 데이터 분석 역량이 있습니다."
```

**❌ 문제**
- 요약만 하고 인사이트 부족
- 실질적인 도움이 되지 않음

---

## 📄 3단계: 목적 추가 – 활용 목적 명확화
**프롬프트**
```
"다음 이력서를 바탕으로 면접관에게 제공할 요약과 강점 3가지, 약점 2가지를 도출해줘.
- 이름: 김지민
- 경력: 마케팅 5년
- 스킬: SNS 광고, 콘텐츠 기획, 데이터 분석
- 주요 프로젝트: 인스타그램 팔로워 200% 증가 캠페인 기획 및 실행"
```

**💡 출력 예시**
```
📄 요약:
김지민 님은 마케팅 분야에서 5년 경력을 가지고 있으며, SNS 광고와 콘텐츠 기획에 강점이 있습니다. 특히 인스타그램 팔로워를 200% 증가시킨 성공적인 캠페인 경험이 있습니다.

💪 강점:
1. SNS 광고 및 캠페인 기획 역량
2. 데이터 기반 마케팅 전략 수립
3. 성과 지향적인 업무 수행 능력

⚠️ 약점:
1. 글로벌 마케팅 경험 부족
2. 최신 마케팅 자동화 툴 활용 경험 제한적
```

---

## 📄 4단계: 포맷 + 역할 부여 – 고도화된 프롬프트
**프롬프트**
```
"너는 전문 헤드헌터입니다. 아래 이력서를 분석하여 면접관에게 제공할 요약, 강점 3가지, 약점 2가지, 추천 질문 3개를 작성해줘.
결과는 Markdown 형식으로 출력해줘.

이력서:
- 이름: 김지민
- 경력: 마케팅 5년
- 스킬: SNS 광고, 콘텐츠 기획, 데이터 분석, GA4
- 주요 프로젝트:
  - 인스타그램 팔로워 200% 증가 캠페인
  - 구글 애널리틱스를 활용한 고객 여정 분석 및 개선"
```

**💡 출력 예시**
```
## 🧾 이력서 분석 – 김지민 후보

### 📄 요약
- 김지민 님은 마케팅 경력 5년의 전문가로 SNS 광고와 콘텐츠 기획에 강점을 가지고 있습니다.
- 인스타그램 팔로워를 200% 증가시킨 캠페인과, 구글 애널리틱스를 활용한 고객 여정 최적화 경험이 있습니다.

### 💪 강점
1. SNS 광고 캠페인 설계 및 실행 역량
2. 데이터 기반 고객 분석 및 인사이트 도출 능력 (GA4 활용)
3. 목표 대비 성과를 달성하는 실행력과 문제 해결력

### ⚠️ 약점
1. 해외 시장 또는 글로벌 캠페인 경험 부족
2. 최신 마케팅 자동화 도구(Marketo, HubSpot) 사용 경험 제한적

### ❓ 추천 면접 질문
1. "구글 애널리틱스를 활용한 고객 여정 분석에서 가장 인사이트가 컸던 부분은 무엇인가요?"
2. "향후 마케팅 자동화 툴을 활용해보고 싶은 영역은 어디인가요?"
3. "팔로워 200% 증가 캠페인 당시 가장 어려웠던 점과 극복 방법은?"
```

---

## 📄 5단계: 확장 – 다양하게 활용하기
### ▶️ 직무별 커스터마이징 가능
- 데이터 분석가용 질문
- 개발자용 스킬 매칭
- 세일즈 직무용 강점 도출

### ▶️ 형식화된 자동화 가능
- Excel → GPT → 분석 리포트 자동화
- Notion + GPT → 인재 데이터베이스 자동 구축

---

## 🎯 💡 교육 포인트
- 프롬프트가 구체적일수록 결과는 정확해진다.
- 역할(Role)을 부여하면 품질이 상승한다.
- 출력 포맷을 명확히 하면 즉시 활용 가능한 자료가 된다.
